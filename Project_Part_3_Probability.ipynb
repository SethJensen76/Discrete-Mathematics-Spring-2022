## Probability

I will be showing the probability of drawing certain cards from a deck, but specifically just Queens, Kings, and then drawing neither. So Queens will be $Q$ which is $\frac{4}{52}$, Kings, $K$ will be $\frac {4}{52}$ and finally, not drawing either will be $N$, which would end up being $\frac{44}{52}$.

King-and-Queen-games-rummy-culture-740x414-1.webp

```
 p =[4/52,4/52,44/52]

 sum(p)

```

Let's see if this will be a compound roll, and we will call this "re-roll" $RR$.

```
ProbabilityRR = 0

for i in [0,2]:
  ProbabilityRR = ProbabilityRR + p[i]

ProbabilityRR
```

The probability of a natural given you get another roll:

$$
P(N|R)=\frac{P(N\cap R)}{P(R)}
$$

```
4/52/ProbabilityRR
```

So this basically shows that if you roll again, you will get it back again all of the time, which is cool to me!

The following is Bayes' formula:

$$
P(N|R) = \frac{P(N)P(R|N)}{P(R)}
$$

```
0.083333333333333333 == (4/52 * 1)/ProbabilityRR
```

Now for a little bit harder calculation. I'm going to convert my `p` by multipling each line by 52, which is the least common multiple of the 3 that I used at the top. I'll call it `p52`

```
p52 = [52*i for i in p]
p =[4/52/9,4/52,44/52]
```

```
from random import choices


choices(['Queens','Kings','Neither'],p52,k=20)
```

```
choices(['Queens','Kings','Neither'],p,k=20)
```

These are actually almost identical, which is really neat to me considering it is all random!

```
DiffCards =  choices(['Queens','Kings','Neither'],p52,k=10000)

countQueens = 0
countKings = 0
countNeither = 0

for i in DiffCards:
  if i =='Queens':
    countQueens += 1
  elif i == 'Kings':
    countKings += 1
  elif i == 'Neither':
    countNeither += 1

estimateProbs = [countQueens/len(DiffCards),countKings/len(DiffCards),countNeither/len(DiffCards)]
```
