## Logic

So, for the three propositions, I am going to use some of what I like to do for fun, in particular, it has to do with me playing Golf. I will label these as 
  * G = PlayGolf 
  * H = HaveFun
  * S = ShootUnder80

```
PlayGolf = True
HaveFun = True
ShootUnder80 = False
```

So the first one that I am going to do is $$S \land G$$
This means I will "Shoot under 80 and play golf."

```
(ShootUnder80) and (PlayGolf)
```

The next that I will do is $$S \lor \lnot H$$
This means that I will either "Shoot under 80, or not have fun."

```
(ShootUnder80) or not(HaveFun)
```

The last one that I will be doing is $$\lnot S \lor G$$
This one means that I will "Not shoot under 80, or play golf."

```
not(ShootUnder80) or (PlayGolf)
```

# Truth Tables

Now I am going to be developing the truth table for
$$ \lnot (S \oplus G) \Rightarrow (G \lor H)$$

First, I am going to go ahead and istall a truth table generator to make things a little easier

```
pip install truth-table-generator
```

Also, in order to get to the desired truth table, I have to do another one first.

```
import ttg
table = ttg.Truths(['G', 'H', 'S'], ['~H', 'S and G', 'S or ~H', 'S => G'], ints = False)
table.as_pandas().style.set_properties(**{'text-align': 'left'}).hide_index()
```

Now for the desired truth table

```
table = ttg.Truths(['G', 'H', 'S'], ['G xor H', '~(G xor H)', 'H or S', '(~(G xor H)) => (H or S)'], ints = False)
table.as_pandas().style.set_properties(**{'text-align': 'left'}).hide_index()
```
