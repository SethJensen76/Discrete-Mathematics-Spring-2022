### Powersets Part 2

I will attempt to do a second part of powersets to see if there is any differences than the first attempt. It would appear that they are nearly identical, only difference being the list format.

```
def powerset(s):
    x = len(s)
    masks = [2**i for i in range(x)]
    for i in range(2**x):
        yield [ss for mask, ss in zip(masks, s) if i & mask]

print(list(powerset(MVP)))
```
